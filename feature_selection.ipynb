### Removing Constant Features

The first feature selection technique involves indicating the constant features. This is done using *Variance Threshold* from sklearn. This works for numeric columns but we can one-hot encode the categorical variables. Let's test this out. Our data set has 3 variables that are constant variables.

```python
# bring in our data set
data = pd.read_csv('data/data_eda.csv')
```

```python
data.head()
```

```python
# these columns have only 1 value so they are constant and can be dropped from the data.
col = list(data.columns)
for col in data[col]:
    unique = len(data[col].unique())
    if unique == 1:
        print(col)
```
